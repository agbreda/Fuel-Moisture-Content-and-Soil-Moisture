```python
import pickle
import numpy as np
from datetime import datetime, timedelta
```

```python
ndays, time = 0, []
for y in range(2001,2019):
    fname = str('upscalemedian_%4.4i.pkl' % y)
    with open(fname, 'rb') as f:
        x = pickle.load(f)
        print('\nYear', y)
        print('Time steps:   ', x[0], x[1].size, x[1][0], x[1][-1])
        print('Lat:   ', x[2].size, x[2][0], x[2][-1])
        print('Lon:   ', x[3].size, x[3][0], x[3][-1])
        print('Data:  ', x[4].shape)
        ndays += x[1].size
        time += [x[1][i] for i in range(x[1].size)]

time = np.array(time)
print('\n\nndays = ', ndays)
print('time.size =', time.size, ';   time[0] =', time[0], ';   time[-1] =', time[-1])
```

```python
lat = x[2]
lon = x[3]
fmc5km = np.zeros((ndays, lat.size, lon.size))
t = 0

for y in range(2001,2019):
    fname = str('upscalemedian_%4.4i.pkl' % y)
    
    with open(fname, 'rb') as f:
        x = pickle.load(f)
        
        for u in range(x[1].size):
            fmc5km[t,:,:] = x[4][u,:,:]
            t += 1
    
    print('Day', t, 'ok!')


# with open('fmc5km_2002_2014.pkl','wb') as f:
#     pickle.dump((time, lat, lon, fmc5km), f)
# ---------------------------------------------------------------------------
# OverflowError                             Traceback (most recent call last)
# <ipython-input-6-a018f5be4267> in <module>
#      17 
#      18 with open('fmc5km_2002_2014.pkl','wb') as f:
# ---> 19     pickle.dump((time, lat, lon, fmc5km), f)

# OverflowError: cannot serialize a bytes object larger than 4 GiB


# The magnific process to create a netCDF file
from netCDF4 import Dataset

# Create the dataset
ds = Dataset('fmc5km_2001_2018.nc','w',format='NETCDF4_CLASSIC')

# Insert dimensions
lat_dim = ds.createDimension('lat', lat.size)
lon_dim = ds.createDimension('lon', lon.size)
time_dim = ds.createDimension('time', time.size)

# Insert variables
#    Dataset.createVariable(<var_id>, <type>, <dimensions>)
times = ds.createVariable('time', np.float32, ('time',))
latitudes = ds.createVariable('latitude', np.float32, ('lat',))
longitudes = ds.createVariable('longitude', np.float32, ('lon',))
fmc_data = ds.createVariable('fmc_median', np.float32, ('time','lat','lon'))

# Including values in the variables
times[:] = time[:]
latitudes[:] = lat[:]
longitudes[:] = lon[:]
fmc_data[:,:,:] = fmc5km[:,:,:]

# Close file
ds.close()
```
